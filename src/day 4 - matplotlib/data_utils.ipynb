### Matplotlib - Synthetic data and ploting

```python
import matplotlib.pyplot as plt
import numpy as np
```

Generate Liner Data

```python
def generate_linear_data(n=100, noise=0.3):
    X = np.random.uniform(-5, 5, n)
    true_y = 2 * X + 1
    y = true_y + np.random.normal(0, noise, n)
    
    return X, y
```

```python
X, y = generate_linear_data(120, .5)
line_y = 2 * X + 1

plt.scatter(X, y, label="data")
plt.plot(sorted(X), sorted(2*X+1), color="red", label="true line")  # simple line plot
for xi, yi, li in zip(X, y, line_y):
    plt.plot([xi, xi], [yi, li], color="gray", alpha=0.3)  # residual: vertical segment
plt.legend()
plt.savefig("./day4-data.png")
```
